# What is the site with the highest job opportunities?

```python
location_of_top = df.location.value_counts() # This statement for show the unique values and counts of times show
name_of_coulmns = ['Europe', 'Uk', 'South east', 'London', 'M4 corridor', 'Switzerland','Germany', 'Cambridge', 'North west', 'Scotland', 'Manchester','Birmingham', 'Oxford', 'France', 'Italy', 'Spain', 'Paris', 'Basel','Portugal'] # names of the columns 
plt.figure(figsize=[28,10]) # This statement for size of the charts
plt.xticks(rotation=90) # This statement for write in x axis or modify
plt.xticks(fontsize = '27', family = 'fantasy',color='black'); # This statement for write in x axis and modify
plt.yticks(fontsize = '25', family = 'fantasy',color='black'); # This statement for write in y axis and modify
sns.barplot(x=name_of_coulmns ,y=location_of_top); # this statement for create a chatr and modify
plt.xlabel('Opportunity sites',fontsize = 35, weight = 'bold',color='black'); # This statement for write in x axis or modify
plt.ylabel('Number of jobs',fontsize = 35, weight = 'bold',color='black'); # This statement for write in x axis or modify
plt.title('The site with the highest job opportunities' , fontsize = 45 ,style='italic',weight='bold') # This statement for write in title and modify
plt.grid() # This statement for drow on the chart by line
plt.savefig('platform')
```

# What is the ratio of different job types?

```python
pd.DataFrame(df.job_type.value_counts()) # This statement for show the unique values and counts of times show 
```

```python
ratio_job_type = df.job_type.value_counts() # This statement for show the values unique and counts of times show 
plt.figure(figsize=[15,5]) # This statement for size of the charts
labels=['Permanent','Contract/Interim','Contract/Temp',' Temporary/Seasonal','Any','Part-Time'] # names of the columns 
explode = [0.1,0,.1,.3,.5,.6] # This statement for keep distance between the offer
plt.pie(ratio_job_type,labels=labels,startangle=90, explode=explode ,autopct='%1.1f%%'); # This statement for create a chatr and modify 
plt.xlabel('Ratio of different job types',fontsize = 20, weight = 'bold',color='black'); # This statement for write in x and axis modify
plt.tight_layout()
```
